```python
import os
import csv
import re
from py2neo import Graph, Node, Relationship

# ======== 1Ô∏è‚É£ Neo4j Aura ËøûÊé•ÈÖçÁΩÆ ========
uri = "neo4j+s://03e29abd.databases.neo4j.io"
username = "neo4j"
password = "50n7LTYnSOMbSuyfbrF2qA7LD8cSNmgFy6B3seFGklo"

graph = Graph(uri, auth=(username, password))
print("‚úÖ Connected to Neo4j Aura")


# ======== 2Ô∏è‚É£ Êï∞ÊçÆË∑ØÂæÑ ========
base_dir = r"E:\finetuning\seven"


# ======== 3Ô∏è‚É£ ÂàõÂª∫ÂîØ‰∏ÄÁ∫¶ÊùüÔºà‰ªÖÊâßË°å‰∏ÄÊ¨°Ôºâ ========
constraints = [
    ("paper_title_unique", "Paper", "title"),
    ("author_name_unique", "Author", "name"),
    ("year_value_unique", "Year", "value"),
    ("keyword_name_unique", "Keyword", "name")
]

for name, label, field in constraints:
    cypher = f"""
    CREATE CONSTRAINT {name} IF NOT EXISTS
    FOR (n:{label}) REQUIRE n.{field} IS UNIQUE
    """
    graph.run(cypher)
print("üîí Constraints created (or already exist)")


# ======== 4Ô∏è‚É£ ÂØºÂÖ•ÂáΩÊï∞ ========
def import_csv_to_neo4j(file_path):
    print(f"\nüìÑ Importing file: {os.path.basename(file_path)}")

    # Ëá™Âä®Ê£ÄÊµãÁºñÁ†Å
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f)
            rows = list(reader)
    except UnicodeDecodeError:
        with open(file_path, 'r', encoding='gbk') as f:
            reader = csv.DictReader(f)
            rows = list(reader)

    count_paper = count_author = count_keyword = 0

    for row in rows:
        # ---- ‰ºòÂÖà‰ΩøÁî® Author full names ----
        authors = row.get("Author full names", "").strip()
        if not authors:  # Â¶ÇÊûú‰∏∫Á©∫ÔºåÂàôÂõûÈÄÄÂà∞‚Äú‰ΩúËÄÖ‚ÄùÂàó
            authors = row.get("‰ΩúËÄÖ", "").strip()

        title = row.get("ÊñáÁåÆÊ†áÈ¢ò", "").strip()
        year = row.get("Âπ¥‰ªΩ", "").strip()
        keywords = row.get("Á¥¢ÂºïÂÖ≥ÈîÆÂ≠ó", "").strip()

        if not title:
            continue  # Ê≤°ÊúâÊ†áÈ¢òÂ∞±Ë∑≥Ëøá

        # === ÂàõÂª∫ Paper ËäÇÁÇπ ===
        paper_node = Node("Paper", title=title)
        graph.merge(paper_node, "Paper", "title")
        count_paper += 1

        # === ‰ΩúËÄÖËäÇÁÇπ ===
        if authors:
            for a in re.split(r'[;,ÔºõÔºå]', authors):
                a = a.strip()
                if a:
                    author_node = Node("Author", name=a)
                    graph.merge(author_node, "Author", "name")
                    graph.merge(Relationship(paper_node, "WRITTEN_BY", author_node))
                    count_author += 1

        # === Âπ¥‰ªΩËäÇÁÇπ ===
        if year:
            year_node = Node("Year", value=year)
            graph.merge(year_node, "Year", "value")
            graph.merge(Relationship(paper_node, "PUBLISHED_IN", year_node))

        # === ÂÖ≥ÈîÆËØçËäÇÁÇπ ===
        if keywords:
            for kw in re.split(r'[;,ÔºõÔºå]', keywords):
                kw = kw.strip()
                if kw:
                    kw_node = Node("Keyword", name=kw)
                    graph.merge(kw_node, "Keyword", "name")
                    graph.merge(Relationship(paper_node, "HAS_KEYWORD", kw_node))
                    count_keyword += 1

    print(f"‚úÖ Done: {count_paper} papers, {count_author} authors, {count_keyword} keywords imported.")


# ======== 5Ô∏è‚É£ ÈÅçÂéÜÊñá‰ª∂Â§π‰∏≠ÁöÑÊâÄÊúâ CSV Êñá‰ª∂ ========
csv_files = [f for f in os.listdir(base_dir) if f.lower().endswith(".csv")]
print(f"\nüì¶ Found {len(csv_files)} CSV files in folder '{base_dir}'")

for file in csv_files:
    full_path = os.path.join(base_dir, file)
    try:
        import_csv_to_neo4j(full_path)
    except Exception as e:
        print(f"‚ö†Ô∏è Failed to import {file}: {e}")

print("\nüéâ All CSV files have been imported successfully into Neo4j!")
```
