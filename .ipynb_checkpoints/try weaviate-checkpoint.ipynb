```python
for m in chunks:
    print(f"sentence retrieved：{m['sentence']}")
    print(f"relative chunk：{m['chunk']}")
    print(f"source paper: {m['title']}")
    refs = m['references']
    figs = m['figs']
    if refs:
        print("references:")
        for k, v in refs.items():
            print(f"[{k}] {v}")
    else:
        print("No reference mentioned")  
    if figs:
        print("figs:")
        for fig in figs:
            figname = fig['img_path']
            print(f"Fig {fig['fig_num']}: {fig['fig_content']}")
            show_image(figname)
    else:
        print("No figs mentioned")  
    print("-" * 40 )
```
